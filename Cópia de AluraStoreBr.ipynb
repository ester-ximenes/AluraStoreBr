### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python

faturamentos = {
    'Loja 1': loja['Preço'].sum(),
    'Loja 2': loja2['Preço'].sum(),
    'Loja 3': loja3['Preço'].sum(),
    'Loja 4': loja4['Preço'].sum()
}
```

```python

import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))
plt.bar(faturamentos.keys(), faturamentos.values(), color='green')
plt.title('Faturamento Total por Loja')
plt.ylabel('R$')
plt.xlabel('Lojas')
plt.tight_layout()
plt.show()
```

# 2. Vendas por Categoria


```python
def vendas_por_categoria(df):
    return df['Categoria do Produto'].value_counts()

vendas_cat_1 = vendas_por_categoria(loja)
vendas_cat_2 = vendas_por_categoria(loja2)
vendas_cat_3 = vendas_por_categoria(loja3)
vendas_cat_4 = vendas_por_categoria(loja4)
```

```python

vendas_cat_1.plot(kind='bar', title='Vendas por Categoria - Loja 1', figsize=(8, 4), color='skyblue')
plt.xlabel('Categoria')
plt.ylabel('Quantidade Vendida')
plt.tight_layout()
plt.show()
```

# 3. Média de Avaliação das Lojas

```python

avaliacoes = {
    'Loja 1': loja['Avaliação da compra'].mean(),
    'Loja 2': loja2['Avaliação da compra'].mean(),
    'Loja 3': loja3['Avaliação da compra'].mean(),
    'Loja 4': loja4['Avaliação da compra'].mean()
}
```

```python

plt.figure(figsize=(8, 5))
plt.plot(list(avaliacoes.keys()), list(avaliacoes.values()), marker='o', color='orange')
plt.title('Média das Avaliações por Loja')
plt.ylabel('Avaliação (1 a 5)')
plt.ylim(0, 5)
plt.grid(True)
plt.tight_layout()
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python

produtos_mais_vendidos_1 = loja['Produto'].value_counts().head(10)  # Exemplo de 10 mais vendidos
produtos_menos_vendidos_1 = loja['Produto'].value_counts().tail(10)  # Exemplo de 10 menos vendidos

# Exibindo o gráfico para os mais vendidos
produtos_mais_vendidos_1.plot(kind='bar', title='Produtos Mais Vendidos - Loja 1', figsize=(8, 4), color='lightcoral')
plt.xlabel('Produto')
plt.ylabel('Quantidade Vendida')
plt.tight_layout()
plt.show()

# Exibindo o gráfico para os menos vendidos
produtos_menos_vendidos_1.plot(kind='bar', title='Produtos Menos Vendidos - Loja 1', figsize=(8, 4), color='lightgreen')
plt.xlabel('Produto')
plt.ylabel('Quantidade Vendida')
plt.tight_layout()
plt.show()
```

# 5. Frete Médio por Loja


```python

fretes = {
    'Loja 1': loja['Frete'].mean(),
    'Loja 2': loja2['Frete'].mean(),
    'Loja 3': loja3['Frete'].mean(),
    'Loja 4': loja4['Frete'].mean()
}
```

```python

plt.figure(figsize=(8, 5))
plt.bar(fretes.keys(), fretes.values(), color='blue')
plt.title('Frete Médio por Loja')
plt.ylabel('R$')
plt.xlabel('Lojas')
plt.tight_layout()
plt.show()
```

Relatório de Análise para o Senhor João: Escolha da Loja a Ser Vendida

Introdução

O objetivo desta análise é auxiliar o Senhor João na decisão sobre qual loja vender, com base em dados fornecidos sobre o faturamento, categorias de produtos, avaliações de clientes, produtos mais e menos vendidos e o frete médio de quatro lojas. Para isso, foram analisadas as seguintes métricas e visualizações, com o intuito de apresentar um panorama claro e objetivo sobre o desempenho de cada loja.

Desenvolvimento

Faturamento Total das Lojas

O faturamento total de cada loja foi calculado somando os valores da coluna "Preço" de todas as vendas. Os resultados são os seguintes:

Loja 1: R$ 1.534.509,12

Loja 2: R$ 1.488.459,06

Loja 3: R$ 1.464.025,03

Loja 4: R$ 1.384.497,58


A Loja 1 foi a que apresentou o maior faturamento, seguida de perto pela Loja 2. As Lojas 3 e 4 ficaram abaixo das demais em termos de faturamento total.

Categorias de Produtos Mais e Menos Vendidas

Analisando as categorias de produtos mais e menos vendidas em cada loja, é possível observar os seguintes padrões:

Loja 1: A categoria de eletrônicos foi a mais vendida, enquanto a categoria de roupas teve o menor número de vendas.

Loja 2: Brinquedos e eletrônicos se destacaram, com alimentos sendo a categoria com menos vendas.

Loja 3: Cosméticos liderou as vendas, enquanto roupas foi a categoria menos procurada.

Loja 4: A categoria de alimentos foi a mais vendida, enquanto cosméticos teve um desempenho inferior.


Essas informações podem ser úteis para o Senhor João decidir qual loja tem maior potencial de acordo com o tipo de produto que ele deseja focar.

Média de Avaliação dos Clientes

A média das avaliações dos clientes para cada loja foi a seguinte:

Loja 1: 3.98

Loja 2: 4.04

Loja 3: 4.05

Loja 4: 3.99


A Loja 3 apresenta a melhor avaliação média, seguida de perto pela Loja 2. A Loja 1 e Loja 4 ficaram abaixo da média, o que pode indicar áreas para melhorias na experiência do cliente.

Produtos Mais e Menos Vendidos

Analisando os produtos mais e menos vendidos em cada loja, foi possível identificar quais itens possuem maior demanda. A Loja 1 teve um produto de eletrônicos como mais vendido, enquanto a Loja 3 teve cosméticos. Já as Lojas 2 e 4 apresentaram uma diversificação maior, com brinquedos e alimentos sendo os produtos mais procurados.

Frete Médio por Loja

O frete médio de cada loja também foi calculado, com os seguintes resultados:

Loja 1: R$ 34,69

Loja 2: R$ 33,62

Loja 3: R$ 33,07

Loja 4: R$ 31,28


A Loja 4 apresentou o menor valor de frete, o que pode ser um atrativo para os clientes que buscam reduzir os custos de envio. Já a Loja 1 apresentou o maior valor de frete, o que pode ser um fator negativo, especialmente se comparado com as outras lojas.

Conclusão

Com base nas análises feitas e nos gráficos gerados, minha recomendação para o Senhor João é vender a Loja 3. A razão principal para essa recomendação é que a Loja 3 apresentou a melhor média de avaliação dos clientes (4.05), indicando um bom nível de satisfação e fidelidade do consumidor. Além disso, o faturamento é significativo e a loja tem um bom desempenho nas categorias de cosméticos, que demonstraram grande demanda.

Embora a Loja 1 tenha o maior faturamento, sua avaliação de clientes é inferior à da Loja 3, o que sugere que, apesar do faturamento, há possíveis problemas de experiência do cliente que poderiam impactar negativamente a fidelização de clientes.

A Loja 4, apesar de ter o frete mais baixo, não apresentou um desempenho tão bom nas vendas e avaliações, o que a coloca em uma posição menos favorável em relação à Loja 3.

Portanto, a Loja 3 se destaca por sua combinação de boa avaliação dos clientes, bom desempenho de vendas e uma categoria de produtos popular. A Loja 3 é a melhor opção para o Senhor João vender, considerando todos os fatores analisados.
